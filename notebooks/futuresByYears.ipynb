```python
import sys
import pandas as pd
import datetime as dt

sys.path.append('..')

from pseudoFrameWork.api import getAnyMOEX
```

```python
futures = pd.read_excel('Классификация фьючерсов.xlsx', sheet_name='окт25')
futures['StarYear'] = futures['Start'].dt.year
futures.sort_values(by='StarYear', inplace=True)
futures.reset_index(drop=True, inplace=True)
futures = futures[futures['StarYear'] >= 2020]


quotes = await getAnyMOEX(
    api_method=r'/iss/history/engines/futures/markets/forts/sessions/rufd/securities.json',
    arguments={'date': '2025-10-01'},
    field='history',
    all=True
)

list = await getAnyMOEX(
    api_method=r'/iss/statistics/engines/futures/markets/forts/series.json',
    arguments={'show_expired': 1},
    field='series',
    all=False
)

quotes = quotes.merge(
    list[['secid', 'asset_code']].rename(columns={'secid': 'SECID'}),
    on='SECID',
    how='left'
)

futures = futures.merge(
    quotes[['asset_code', 'CLOSE']].dropna().groupby('asset_code').min().reset_index(),
    left_on='ASSETCODE',
    right_on='asset_code',
    how='left'
).drop('asset_code', axis=1)

futures.loc[futures['ASSETCODE'] == 'KZT', 'CLOSE'] = 13.36
futures.loc[futures['ASSETCODE'] == 'BYN', 'CLOSE'] = 27.47
futures.loc[futures['ASSETCODE'] == 'TENCENT', 'CLOSE'] = 670.1

futures.drop(['Mini', 'Perpetuity'], axis=1, inplace=True)

futures.loc[:, ['GO1', 'GO2', 'GO3', 'CLOSE']] = futures.loc[:, ['GO1', 'GO2', 'GO3', 'CLOSE']].round(2)
futures.loc[:, ['GO1', 'GO2', 'GO3', 'CLOSE']] = futures.loc[:, ['GO1', 'GO2', 'GO3', 'CLOSE']].astype(str).replace('\.', ',', regex=True)
futures.loc[:, ['GO1', 'GO2', 'GO3', 'CLOSE']] = futures.loc[:, ['GO1', 'GO2', 'GO3', 'CLOSE']].replace('nan', '0')
futures['Start'] = futures['Start'].dt.strftime('%d.%m.%y')
futures['Finish'] = futures['Finish'].dt.strftime('%d.%m.%y')

futures = futures.astype(str)
futures.rename(columns={
    'ASSETCODE': 'Код',
'GROUPTYPE': 'Группа',
'Start': 'Начало обращения',
'Finish': 'Исполнение',
'CONTRACTNAME': 'Наименование',
'class1': 'Секция',
'class2': 'Базовые',
'GO1': '*ГО1',
'GO2': '*ГО2',
'GO3': '*ГО3',
'CLOSE': '**Цена',
}, inplace=True)
```
